# Multi-Disease Prediction EDA

This notebook contains basic exploratory data analysis (EDA) for the datasets used in the Multi-Disease Prediction System.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set style
sns.set_theme(style="whitegrid")
```

## 1. Heart Disease Dataset

```python
if os.path.exists('../data/heart.csv'):
    heart_df = pd.read_csv('../data/heart.csv')
    display(heart_df.head())
    # Target distribution
    target_col = 'target' if 'target' in heart_df.columns else 'class'
    sns.countplot(data=heart_df, x=target_col)
    plt.title('Heart Disease Target Distribution')
    plt.show()
else:
    print("Dataset not found. Run fetch_data.py first.")
```
